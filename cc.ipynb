```python
import pandas as pd 
import numpy as np
uname = [ 'user id', 'age', 'gender', 'occupation', 'zip code']
users = pd.read_csv('C:\\Users\\vimal\\OneDrive\\Documents\\ml-100k\\u.user',sep= '|',names = uname)
rname = ['user id', 'item id', 'rating', 'timestamp']
ratings = pd.read_csv('C:\\Users\\vimal\\OneDrive\\Documents\\ml-100k\\u.data', sep= '\t', names = rname)
users_df = users.loc[:, ['user id', 'gender']]
ratings_df = ratings.loc[:, ['user id', 'rating']]
rating_df = pd.merge(users_df, ratings_df)
rating_df
```

```python
rating_df.groupby('gender').rating.std()
```

```python
df1 = rating_df.groupby(['user id', 'gender']).apply(np.mean)
df1.groupby('gender').rating.std()
```
